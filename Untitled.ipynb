```python
# Get the current directory
current_dir = os.getcwd()

# Construct the relative path to the CSV file
csv_file_path = os.path.join(current_dir, 'Python_Amazon_Sales_Analysis-main', 'Amazon Sale Report.csv')

# Read the CSV file
df = pd.read_csv(csv_file_path, encoding='unicode_escape')
```

```python
df['ship-postal-code'] = df['ship-postal-code'].astype('int')
```

```python
ax=sns.countplot(x='Size',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Size'] ,as_index = False)['Qty'].sum().sort_values(by= 'Qty', ascending = False)
```

```python
S_Qty = df.groupby(['Size'] ,as_index = False)['Qty'].sum().sort_values(by= 'Qty', ascending = False)
sns.barplot(x = 'Size' , y ='Qty', data=S_Qty)
```

```python
plt.figure(figsize = (10,5))
ax=sns.countplot(x = 'Courier Status' , hue = 'Status', data=df)
plt.show()
```

```python
df['Size'].hist()
```

```python
df['Category'] = df['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize = (10,5))
plt.hist(column_data, bins =30 , edgecolor ='black')
plt.xticks(rotation= 90)
plt.show()
```

```python
B2B_Check = df['B2B'].value_counts()
plt.pie(B2B_Check , labels = B2B_Check.index,autopct = '%1.1f%%', wedgeprops=dict(width=0.8))
plt.show()
```

```python
x_data = df['Category']
y_data = df['Size']
plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show()
```

```python
plt.figure(figsize= (12,6))
ad = sns.countplot(data = df, x = 'ship-state')
for bars in ad.containers:
    ad.bar_label(bars, rotation =90, size = 8)
    
    
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Ship State Plot')
plt.xticks(rotation = 90)
plt.show()
```

```python
top_10_states = df['ship-state'].value_counts().head(10)
plt.figure(figsize= (12,6))
ad = sns.countplot(data = df[df['ship-state'].isin(top_10_states.index)], x = 'ship-state')
for bars in ad.containers:
    ad.bar_label(bars, rotation =0, size = 8)
    
    
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Ship State Plot')
plt.xticks(rotation = 45)
plt.show()

```
